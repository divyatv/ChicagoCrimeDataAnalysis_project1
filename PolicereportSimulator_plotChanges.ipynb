```python
###Import required libraries####################################
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# import seaborn as sns
import os
import time
```

```python
### Create a new dataframe which will have crime types and years with their counts for each
print("Starting the read the file this may take a while, thanks for your patience!.........", time.asctime(time.localtime(time.time())))
file = os.path.join('Resources','Crimes_2001_to_present.csv')
crimedata_selected = pd.read_csv(file)
#file_to_load = "crime_datadf_selectedtype.csv"

print(" Completed file reading. It took", time.asctime(time.localtime(time.time())) ,"time to read.")


```

```python

summarydf=crimedata_selected.groupby(['Primary Type','Year'])['ID'].count()
```

```python

##### Create two list to make sure you cover the crime types and years CHICAGO POLICE department uses
crimetypelist=['HOMICIDE', 'THEFT','BATTERY','BURGLARY','MOTOR VEHICLE THEFT']
yearlist=[2015, 2016, 2017, 2018, 2019]
count=[]

newdf=pd.DataFrame(columns=yearlist)
newdf['Crimetype']=crimetypelist

summarydict={2015: [], 2016:[], 2017:[], 2018:[], 2019:[], 'Crimetype':[]}

###### Fill the dataframe #################################
for crimetype in crimetypelist:
    summarydict['Crimetype'].append(crimetype)

    for year in yearlist:
        #newdf[year]=summarydf[crimetype][year]
        summarydict[year].append(summarydf[crimetype][year])

newdf=pd.DataFrame(summarydict)
#print(summarydict)
#snewdf.to_csv("newdf.csv", index = None, header=True)
newdf.head(10)  
```
